```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
df_threads = pd.read_csv('lab3-stats_threads_new.csv')
```

```python
df_threads
```

```python
df_process = pd.read_csv('lab3-stats_processed_new.csv')
df_process
```

```python
plt.plot(df_process['Average Time (ms)'], label='processes')
plt.plot(df_threads['Average Time (ms)'], label='threads')
plt.title('Average Time for same buffer/producers/consumers')
plt.legend()
plt.show()
```

```python
plt.plot(df_process['Standard Deviation (ms)'], label='processes')
plt.plot(df_threads['Standard Deviation (ms)'], label='threads')
plt.title('Standard Deviation for same buffer/producers/consumers')
plt.legend()
plt.show()
```
